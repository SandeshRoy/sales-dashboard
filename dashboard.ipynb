```python
import pandas as pd

df = pd.read_csv("superstore.csv", encoding='latin1')

print(df.columns)
```

```python
customer_sales = df.groupby('Customer ID')['Sales'].sum()

df['Customer Segment Value'] = df['Customer ID'].map(customer_sales)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

df.groupby('Order Date')['Sales'].sum().plot(title="Sales Trend")
```

```python
customer_sales = df.groupby('Customer ID')['Sales'].sum()

df['Customer Segment Value'] = df['Customer ID'].map(customer_sales)
```

```python
def segment(x):
    if x > 5000:
        return "High Value"
    elif x > 2000:
        return "Medium Value"
    else:
        return "Low Value"

df['Customer Type'] = df['Customer Segment Value'].apply(segment)
```

```python
df['Customer Type'].value_counts().plot(kind='bar', title="Customer Segmentation")
```

```python
from prophet import Prophet

df_prophet = df[['Sales']].copy()
df_prophet['ds'] = pd.date_range(start='2020-01-01', periods=len(df))
df_prophet = df_prophet.rename(columns={'Sales': 'y'})

model = Prophet()
model.fit(df_prophet)

future = model.make_future_dataframe(periods=30)
forecast = model.predict(future)

model.plot(forecast)
```

```python
top_products = df.groupby('Product Name')['Sales'].sum() \
                 .sort_values(ascending=False).head(10)

top_products.plot(kind='barh', title="Top 10 Products")
```

```python
import matplotlib.pyplot as plt

top_products = df.groupby('Product Name')['Sales'].sum() \
                 .sort_values(ascending=True).tail(10)

plt.figure(figsize=(10,6))
top_products.plot(kind='barh', color='skyblue')

plt.title("Top 10 Products by Sales", fontsize=14)
plt.xlabel("Sales")
plt.ylabel("Product")

plt.tight_layout()
plt.show()
```
